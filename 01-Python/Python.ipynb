**Time**

The explanation of each of these functions from the time module in Python:

- `time.time()`: Returns the current time in seconds since the Epoch (January 1, 1970, 00:00:00 UTC) as a floating-point number.

- `time.sleep(secs)`: Suspends execution of the current thread for the given number of seconds (can be a floating-point value).

- `time.localtime([secs])`: Converts a time expressed in seconds since the Epoch (or the current time if secs is not provided) into a struct_time object representing local time (i.e., the time zone is considered).

- `time.asctime([t])`: Converts a struct_time object or a tuple containing 9 elements representing a time to a string representing the time in the format "Day Mon dd hh:mm:ss year". If t is not provided, it uses the current local time.

- `time.strptime(date_string, format)`: Parses a string representing a time in a given format and returns a struct_time object representing that time. The format parameter specifies the expected format of the input string.

```python
import time;

ticks = time.time()
print("Number of ticks since 12:00am, January 1, 1970:", ticks)

time.sleep(0.1)

ticksagain = time.time()
print("Again, Number of ticks since 12:00am, January 1, 1970:", ticksagain)
print("Difference:", (ticksagain-ticks))

a = time.localtime()
print("a:", a, ", type:", type(a))

b = time.asctime(a) #ASCII time
print("b:", b, ", type:", type(b))

d = time.strptime("30 Nov 22", "%d %b %y")  #string parse time, day,abbreviated month,year
print("d:", time.asctime(d))
```

**JSON**

The explanation of each of these functions from the json module in Python:

- `json.loads(x)`: Takes a JSON string x and converts it to a Python object. The resulting object can be a list, dictionary, string, number, or boolean value.

- `json.dumps(y)`: Takes a Python object y and converts it to a JSON string. The resulting string represents the object as a JSON-encoded value.

- `json.dumps(y, indent=4, separators=(",", ": "), sort_keys=True)`: This is a variation of json.dumps() function that allows you to customize the JSON output. The indent parameter is used to specify the number of spaces for indentation. The separators parameter is used to customize the separators between the key-value pairs in the JSON output. The sort_keys parameter, when set to True, will sort the keys in the output in alphabetical order.

```python
import json

x = '{ "name": "Ahmed Ali", "age": 15, "address": { "country":"Egypt","city":"Cairo" } }'

y = json.loads(x)
print(type(y))
print(y['name'])
print(y['address']['city'])

a = json.dumps(y)
print(type(a))
print(a)

a = json.dumps(y, indent = 4, separators=(",", ": "), sort_keys=False)
print(a)
```

**Regular Expression**

- The function **re.search("^The", text)** is used to search a string text for a match with a regular expression pattern ^The. The ^ character is a special character in regular expressions that indicates the start of a line or string. Therefore, this pattern will match any line or string that starts with the word "The". The search function returns a match object if the pattern is found, and None if there is no match. The match object contains information about the location and contents of the match, and can be used to extract the matched substring or perform other operations on the matched text.

```python
import re

text = "The rain in Spain"

print("Begin with 'The':", re.search("^The", text))
print("Begin with 'The':", re.search("^The.*", text))

print("Another way for begin with '\The':", re.search(r"\AThe", text))

print("Ends with 'Spain':", re.search("Spain$", text))

print("Another way for ends with 'Spain':", re.search(r"Spain\b", text))

print("Begin with 'The' and ends with 'Spain':", re.search("^The.*Spain", text))
```

```python
print("Contains 'Spain':", re.search("Spain", text))
print("Contains 'spain':", re.search("spain", text, re.IGNORECASE))
print("Contains 'he' then anything then 'Sp':", re.search("he.*Sp", text))
```

```python
text = "soon, noon, room "

print("Contains s followed by any two letters then n:", re.search("s.{2}n", text))

print("Contains s|n|r followed by any two letters then n|m:", re.search("[snr].{2}[nm]", text))

print("Find all, contains s|n|r followed by any two letters then n|m:", re.findall("[snr].{2}[nm]", text))

x = re.search("[snr].{2}[nm]", text)
print(x.span(), x.group())
```

## PYTHON, CLASSES
**Define Class: BasicStudent**


* **Special methods** have names that begin and end with double underscores (__). These methods are also called "magic methods" or "dunder methods" (short for "double underscore"). They are used to define behavior for built-in operations like arithmetic, comparisons, and iteration, among others. Examples of special methods include __add__, __eq__, __iter__, and __str__.


* **Public methods** have names that don't begin with double underscores. These methods are part of the public interface of a class and can be called from outside the class. They are used to perform operations or expose functionality to users of the class. Public methods are often documented in the class's docstrings. Examples of public methods include add_item, remove_item, and get_size.


* **Private methods** have names that begin with double underscores but don't end with them (e.g. __my_private_method). These methods are intended to be used only within the class and are not part of the public interface. They are used to encapsulate implementation details and separate them from the public interface. Private methods are often used to break down complex tasks into smaller, more manageable pieces. It's worth noting that while private methods are not part of the public interface, they can still be called from other methods within the same class.

```python
class BasicStudent(object):
    def __init__(self, first_name, last_name): # This is the constructor method
        self.first_name = first_name; 
        self.last_name = last_name;
        print(self.__class__.__name__, "constructed AA")

    def __del__(self):  # This is the destructor method
        print(self.__class__.__name__, "destroyed")

    def __str__(self):
        return self.name()
  
    def __private_method(self):   # Private method
        print("This is a private method.")

    def name(self):             # Public method
        self.__private_method()
        return (self.first_name + ' ' + self.last_name) + "AA";
```

```python
student = BasicStudent('ahmeD', 'salem');
print(student.name())
print(student)

# student.__private_method()

del(student)
```

## Inheritance

```python
class Student(BasicStudent):  #Student class inherits from the BasicStudent class
  def __init__(self, first_name, last_name, age):
    super(Student, self).__init__(first_name, last_name)
    self.age = age
    print(self.__class__.__name__, "constructed BB")

  def name(self):   # overriding
    return (self.first_name + ' ' + self.last_name).title()

  def info(self):
    return self.name() + ', '+ str(self.age) + ' year';

  def __del__(self):
    print(self.__class__.__name__, "destroyed")
```

```python
student = Student('ahmeD', 'salem', 18)
print(student.name())
print(student.info())
print(student)
```

**Define Iterator: Primes**

```python
class Primes:
  def __init__(self, max):
    self.count = max

  def __iter__(self):
    self.prime = 2
    return self

  def __is_prime(self, test):
    x = test - 1
    while(x>1):
      if((test%x)==0):
        return False
      x -= 1
    return True

  def __next__(self):
    x = self.prime
    self.prime += 1
    self.count -= 1
    if(self.count==0):
      raise StopIteration;
    while(not self.__is_prime(self.prime)):
        self.prime += 1
    return x


```

```python
primes = Primes(10)
myiter = iter(primes)
for x in myiter:
  print(x)
```

## iloc
- Integer Location Based Indexing: `iloc` is used for positional indexing, i.e., selecting data by integer-based positions (like in a list).
- Syntax: `df.iloc[start:stop]` selects rows from the `start` index to the `stop-1` index (exclusive).
- **Example:** `df.iloc[1:5]` will select rows starting from position 1 up to (but not including) position 5. If the DataFrame has fewer than 5 rows, it will return rows up to the end of the DataFrame.

## loc
- Label Based Indexing: `loc` is used for label-based indexing, i.e., selecting data by row and column labels.
- Syntax: `df.loc[start_label:stop_label]` selects rows from the start_label to the stop_label (inclusive).
- **Example:** `df.loc[1:5]` will attempt to select rows with index labels from 1 to 5. This can result in a KeyError if the DataFrame does not have labels 1 to 5

# DataFrame, Pipes

```python
def removena(df):
    return df.dropna()

def fixdate(df):
    dfc = df.copy()
    dfc['Date'] = pd.to_datetime(dfc['Date'])
    return dfc
    
def removeduplicates(df):
    dfc = df.copy()
    dfc.drop_duplicates()
    return dfc
```

```python
df = pd.read_csv('./data/wdata.csv')
df
```

```python
dfc = df.pipe(removena).pipe(fixdate).pipe(removeduplicates)
dfc
```

```python
def clean(df):
    dfc = df.copy()
    dfc = dfc.dropna()
    dfc['Date'] = pd.to_datetime(dfc['Date'])
    dfc.drop_duplicates()
    return dfc
    
def modify(df, value):
    dfc = df.copy()
    dfc['Final'] = value
    return dfc
```

```python
dfc = df.pipe(clean).pipe(modify, 100)
dfc
```

```python
import numpy as np
from matplotlib.patches import Polygon

def func(x):
    return (x**2)

a, b = 50,100  # integral limits

x = np.linspace(0, 200,1000)
y = func(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'b', linewidth=2)
ax.set_ylim(bottom=0)

# Make the shaded region
ix = np.linspace(a, b)
iy = func(ix)
verts = [(a, 0), *zip(ix, iy), (b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)


ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.set_xticks([a, b], labels=['$a$', '$b$'])
ax.set_yticks([])

plt.show()
```

```python
# color map
alpha = 0.7
phi_ext = 2 * pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return ( + alpha - 2 * cos(phi_p)*cos(phi_m) - 
        alpha * cos(phi_ext - 2*phi_p))

phi_m = linspace(0, 2*pi, 100)
phi_p = linspace(0, 2*pi, 100)
X,Y = meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T

fig, ax = plt.subplots()

p = ax.pcolor(X/(2*pi), Y/(2*pi), Z, 
              cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)
```
